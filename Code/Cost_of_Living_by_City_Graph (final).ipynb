```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Reading in files to graph
# Reading in DataSci data
datasci_data = "../Resources/data_science_marker.csv"
df = pd.read_csv(datasci_data)
# Reading in DataAnalyst data
dataanalyst_data = "../Resources/data_analyst_marker.csv"
df_2 = pd.read_csv(dataanalyst_data)

# Reading in Living Wage data
livingwage_data = "../Resources/Cost of Living by City_with LivingWage.csv"
df_3 = pd.read_csv(livingwage_data)
df_3.dropna(inplace=True)
df_2.head(3)
```

```python
# Bar Chart to visualize Salaries compared to COL using living wages
# x_axis = np.arange(40000, 200000, 20000)

# creating bars and converting to float for y_axis

ds_sal = df['Value']
da_sal = df_2['Value']
living_wage = df_3['livingWage']
city = df_2['city']

bar_df = pd.DataFrame({"City" : city,
         "Data Scientist": ds_sal,
         "Data Analyst": da_sal,
          "Living Wage" : living_wage})

color = ["navy","gold","grey"]
graph = bar_df.plot(kind="bar", x="City", rot=45, fontsize=10, figsize=(12, 6), title="Mean Salaries Compared to Living Wage", legend=False, color=color)
graph.set(ylabel="Living Wage")
plt.legend(loc="upper left")
plt.tight_layout()

# Saves an image of our chart so that we can view it in a folder
plt.savefig("../Output/CostofLiving_Graph.png")
plt.show()
```
