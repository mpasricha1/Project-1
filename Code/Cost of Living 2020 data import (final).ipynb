```python
import pandas as pd
import os
```

```python
# pulling in COL file into df
col_df = pd.read_csv("../Resources/cost_of_living_2020.csv")
col_df.head(10)
```

```python
# creating new df dropping the states that do not contain the cities being analyzed by Muse

city10_df = col_df.loc[(col_df["State"] == "Georgia") 
                       | (col_df["State"] == "Massachusetts") 
                       | (col_df["State"] == "Illinois") 
                       | (col_df["State"] == "Texas") 
                       | (col_df["State"] == "California") 
                       | (col_df["State"] == "New York") 
                       | (col_df["State"] == "Pennsylvania") 
                       | (col_df["State"] == "Washington") 
                       | (col_df["State"] == "Washington DC")] 

                      
city10_df 
```

```python
# exporting to csv file
city10_df.to_csv("../Output/Cost of Living by City.csv", index=False, header=True)
```
