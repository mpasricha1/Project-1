```python
# Dependencies and set up
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File path and read csv
rankCOLFile = "../Output/rankCOL.csv"
rankCOL_df = pd.read_csv(rankCOLFile)
rankCOL_df = rankCOL_df[["Cities","Rank COL"]]
rankCOL_df
```

```python
# data science pull second csv
rankDSsalaryFile = "../Output/rankDSsalary.csv"
rankDSSalary_df = pd.read_csv(rankDSsalaryFile)
rankDSSalary_df=rankDSSalary_df[["Cities","Rank DS Salary"]]
rankDSSalary_df
```

```python
# merge csv on cities
mergeDS1 = pd.merge(rankCOL_df, rankDSSalary_df, on="Cities", how="left")
mergeDS1
```

```python
# pull 3rd csv
rankDSYOYFile = "../Output/rankDSYOY.csv"
rankDSYOY_df = pd.read_csv(rankDSYOYFile)
rankDSYOY_df=rankDSYOY_df[["Cities","Rank DS Salary YOY"]]
rankDSYOY_df
```

```python
# merge on cities
mergeDS2 = pd.merge(mergeDS1, rankDSYOY_df, on="Cities", how="left")
mergeDS2
```

```python
# create alternative final score
altDSRanking = pd.merge(mergeDS1, rankDSYOY_df, on="Cities", how="left")
altDSRanking["Alt Final Score"] = altDSRanking["Rank COL"] + altDSRanking["Rank DS Salary"] + altDSRanking["Rank DS Salary YOY"]
altDSRanking = altDSRanking.sort_values("Alt Final Score")
altDSRanking
```

```python
# create alternative final ranking
altDSRanking['Alt Final Ranking'] = altDSRanking['Alt Final Score'].rank()
altDSRanking
```

```python
# create df visual for powerpoint
presentAltDS = altDSRanking[["Cities", "Alt Final Ranking"]]
presentAltDS
```

```python
# pull 4th csv
rankDSPostFile = "../Output/rankDSPost.csv"
rankDSPost_df = pd.read_csv(rankDSPostFile)
rankDSPost_df
```

```python
# eliminate flexible/remote locations
rankDSPostCity = rankDSPost_df[rankDSPost_df["Cities"] != "Flexible / Remote"]

# grab necessary columns for merge
rankDSPostCity = rankDSPostCity[["Cities", "Rank DS Posts"]]
rankDSPostCity 
```

```python
# merge
mergeDS3 = pd.merge(mergeDS2, rankDSPostCity, on="Cities", how="left")
mergeDS3
```

```python
# pull 5th csv
rankPythonFile = "../Output/rankPythonCount.csv"
rankPython_df = pd.read_csv(rankPythonFile)
rankPython_df =rankPython_df [["Cities","Rank Python"]]
rankPython_df 
```

```python
# merge
mergeDS4 = pd.merge(mergeDS3, rankPython_df , on="Cities", how="left")
mergeDS4
```

```python
# calculate final scores and sort
mergeDS4["Final Score"] = mergeDS4["Rank COL"] + mergeDS4["Rank DS Salary"] + mergeDS4["Rank DS Salary YOY"] + mergeDS4["Rank DS Posts"] + mergeDS4["Rank Python"]
mergeDS4 = mergeDS4.sort_values("Final Score")
mergeDS4
```

```python
# create final ranking
mergeDS4['DS Final Ranking'] = mergeDS4['Final Score'].rank()
mergeDS4
```

```python
# push to csv file
outFile = "../Output/finalDSRank.csv"
mergeDS4.to_csv(outFile)
```

```python
# df visual for ppt
presentDS = mergeDS4[["Cities", "DS Final Ranking"]]
presentDS
```
